```python
import plotly.express as px
import pandas as pd
from scipy.spatial import distance
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Data4.csv',index_col=0)
```

```python
df.head()
```

```python
fig = px.scatter_3d(df, x='X1', y='X2', z='X3',
              color='Class')
fig.update_layout(
    title="Original Classes",
    )
fig.show()
```

```python
X = df[["X1","X2","X3"]] #
y = df["Class"]
```

```python
clusters = DBSCAN().fit(X)
```

```python

df["predicted"] = [ 1 if i == 0  else 2 for i in clusters.labels_]
df.head()
```

```python
fig = px.scatter_3d(df, x='X1', y='X2', z='X3',
              color='predicted')
fig.update_layout(
    title="Predicted Classes using DBSCAN",
    )
fig.show()
```

```python
score = accuracy_score(df["Class"],df["predicted"])
print("Accuracy of result obtained by DBSCAN clustering on given dataset: %4.2f %% "%(score*100))
```
